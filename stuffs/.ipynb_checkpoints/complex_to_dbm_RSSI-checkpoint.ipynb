```python
import matplotlib
matplotlib.use('nbagg')
# %matplotlib notebook
import matplotlib.pyplot as plt
%load_ext autoreload
%autoreload 2

import numpy as np
import scipy
from scipy import signal
```

```python
iq = scipy.fromfile(open('../data/phantom2/uksetaga_f2405350000.00_fs2181818.18_rx6_30_0_band2000000.00.iq'), dtype=scipy.complex64)
iq = iq[500000:507000]

iq /= 2048 * 2 * 2
# https://se.mathworks.com/help/dsp/ref/dbconversion.html
dbm = 10*np.log10(abs(iq)**2/50) + 30 # 50 is ohms

dbm -= 6 + 30 # compensate for bladeRF GAINS
dbm = signal.decimate(dbm, 32, zero_phase=True)
plt.plot(dbm)
plt.xlabel('Samples (decimated)')
plt.ylabel('RSSI (dBm)')
plt.show()
```
